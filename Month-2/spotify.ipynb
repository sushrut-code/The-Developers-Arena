```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_final, x='artist_popularity', y='track_popularity', 
                hue='explicit', alpha=0.6, palette='coolwarm')

# Add a reference line
plt.plot([0, 100], [0, 100], '--', color='gray')

plt.title('The "Viral" Effect: Artist vs. Track Popularity')
plt.xlabel('Artist Popularity (Fame)')
plt.ylabel('Track Popularity (Success)')
plt.legend(title='Explicit')
plt.show()
```

```python
from scipy import stats

# 1. Separate the groups
explicit_pop = df_final[df_final['explicit'] == True]['track_popularity']
clean_pop = df_final[df_final['explicit'] == False]['track_popularity']

# 2. Run the test
t_stat, p_val = stats.ttest_ind(explicit_pop, clean_pop)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_val:.5f}")

# Interpretation logic for your notebook:
if p_val < 0.05:
    print("Result: Statistically Significant difference!")
else:
    print("Result: No significant difference found.")
```
