```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("kidney_dataset.csv")
df
```

```python
#CKD-chronic kidney disease
#bun-blood urea nitrogen-elevated_levels indicate impaired kidney function
#cretinine- blood cretinine level- higher the level, worse the kidney function
#urine_output=amount of urine produced by the kidneys. Low urine output may indicate kidney dysfunction
#age- age of the patient(18-90)- older age is associated with increased risk of kidney disease
#protein_in_urine= high levels of protein in urine may indicate kidney damage
#medication= type of medication the patient is taking- some medications can affect kidney function.
```

```python
df['Urine_Output_in_liters']=df['Urine_Output']/1000
```

```python
df['CKD_Status'].value_counts()
sns.countplot(x='CKD_Status',data=df)
plt.title("Distribution of CKD Status")
plt.show()
```

```python
df['Diabetes'].value_counts()
sns.countplot(x='Diabetes',data=df)
plt.title("Distribution of Diabetes")
plt.show()
```

```python
df['Hypertension'].value_counts()
sns.countplot(x='Hypertension',data=df)
plt.title("Distribution of Hypertension")
plt.show()
```

```python
#given chronic kidney disease status, how many have diabetes
df[df['CKD_Status']==1]['Diabetes'].value_counts()
#percent of ckd patients having diabetes
df[df['CKD_Status']==1]['Diabetes'].value_counts(normalize=True)*100
sns.countplot(x='Diabetes',hue='CKD_Status',data=df)
plt.title("Diabetes vs CKD Status")
plt.show()
```

```python
# given chronic kidney disease status, how many have hypertension
df[df['CKD_Status']==1]['Hypertension'].value_counts()
#percent of ckd patients having hypertension
df[df['CKD_Status']==1]['Hypertension'].value_counts(normalize=True)*100
sns.countplot(x='Hypertension',hue='CKD_Status',data=df)
plt.title("Hypertension vs CKD Status")
plt.show()
```

```python
#given chronic kidney disease status, how many have both diabetes and hypertension
df[(df['CKD_Status']==1) & (df['Diabetes']==1) & (df['Hypertension']==1)].shape[0]
#percent of ckd patients having both diabetes and hypertension
(df[(df['CKD_Status']==1) & (df['Diabetes']==1) & (df['Hypertension']==1)].shape[0]/df[df['CKD_Status']==1].shape[0])*100
sns.countplot(x='CKD_Status',hue='Diabetes',data=df[df['Hypertension']==1])
plt.title("CKD Status vs Diabetes among Hypertensive Patients")
plt.show()
```

```python
df.head()
```

```python
#average Cretinine levels for CKD and non-CKD patients(mg/dL)
df.groupby('CKD_Status')['Creatinine'].mean()
sns.boxplot(x='CKD_Status',y='Creatinine',data=df)
plt.title("Creatinine Levels by CKD Status")
plt.show()
```

```python
# average BUN levels for CKD and non-CKD patients(mg/dL)
df.groupby('CKD_Status')['BUN'].mean()
sns.boxplot(x='CKD_Status',y='BUN',data=df)
plt.title("BUN Levels by CKD Status")
plt.show()
```

```python
# average GFR levels for CKD and non-CKD patients(mL/min/1.73 mÂ²)
df.groupby('CKD_Status')['GFR'].mean()
sns.boxplot(x='CKD_Status',y='GFR',data=df)
plt.title("GFR Levels by CKD Status")
plt.show()
```

```python
# average Urine_Output levels for CKD and non-CKD patients(in liters)
df.groupby('CKD_Status')['Urine_Output_in_liters'].mean()
sns.boxplot(x='CKD_Status',y='Urine_Output_in_liters',data=df)
plt.title("Urine Output Levels by CKD Status")
plt.show()
```

```python
# average age for CKD and non-CKD patients(years)
df.groupby('CKD_Status')['Age'].mean()
sns.boxplot(x='CKD_Status',y='Age',data=df)
plt.title("Age by CKD Status")
plt.show()
```

```python
df['Age'].plot(kind='kde',title='Age Distribution')
```

```python
# average protein_in_urine levels for CKD and non-CKD patients(mg/dL)
df.groupby('CKD_Status')['Protein_in_Urine'].mean()
sns.boxplot(x='CKD_Status',y='Protein_in_Urine',data=df)
plt.title("Protein in Urine Levels by CKD Status")
plt.show()
```

```python
# average water_intake levels for CKD and non-CKD patients(liters/day)
df.groupby('CKD_Status')['Water_Intake'].mean()
sns.boxplot(x='CKD_Status',y='Water_Intake',data=df)
plt.title("Water Intake by CKD Status")
plt.show()
```

```python
#distribution of medication
df['Medication'].value_counts()
sns.countplot(x='Medication',data=df)
plt.title("Distribution of Medication")
plt.show()
```

```python
# distribution of medication among CKD patients
df[df['CKD_Status']==1]['Medication'].value_counts()
sns.countplot(x='Medication',data=df[df['CKD_Status']==1])
plt.title("Distribution of Medication among CKD Patients")
plt.show()
```

```python
# correlation matrix
numeric_cols=df.select_dtypes(include=np.number).columns
corr_matrix=df[numeric_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt=".2f")
```
